**I. Starting, ending, and middle n-grams**

```
words3=[x for x in words if len(x)>2]
words4=[x for x in words if len(x)>3]
words5=[x for x in words if len(x)>4]

end=[x[-1] for x in words]
end3=[x[-3:] for x in words3]
end4=[x[-4:] for x in words4]
end5=[x[-5:] for x in words5]

dictend1 = dict()
for i in end:
  dictend1[i] = dictend1.get(i, 0) + 1
dictend1={k: v for k, v in sorted(dictend1.items(), key=lambda item: item[1],reverse=True)}
dictend3 = dict()
for i in end3:
  dictend3[i] = dictend3.get(i, 0) + 1
dictend3={k: v for k, v in sorted(dictend3.items(), key=lambda item: item[1],reverse=True)}
dictend4 = dict()
for i in end4:
  dictend4[i] = dictend4.get(i, 0) + 1
dictend4={k: v for k, v in sorted(dictend4.items(), key=lambda item: item[1],reverse=True)}
dictend5 = dict()
for i in end5:
  dictend5[i] = dictend5.get(i, 0) + 1
dictend5={k: v for k, v in sorted(dictend5.items(), key=lambda item: item[1],reverse=True)}
```

```

start=[x[0] for x in words]
start3=[x[:3] for x in words3]
start4=[x[:4] for x in words4]
start5=[x[:5] for x in words5]

dictstart1 = dict()
for i in start:
  dictstart1[i] = dictstart1.get(i, 0) + 1
dictstart1={k: v for k, v in sorted(dictstart1.items(), key=lambda item: item[1],reverse=True)}
dictstart3 = dict()
for i in start3:
  dictstart3[i] = dictstart3.get(i, 0) + 1
dictstart3={k: v for k, v in sorted(dictstart3.items(), key=lambda item: item[1],reverse=True)}
dictstart4 = dict()
for i in start4:
  dictstart4[i] = dictstart4.get(i, 0) + 1
dictstart4={k: v for k, v in sorted(dictstart4.items(), key=lambda item: item[1],reverse=True)}
dictstart5 = dict()
for i in start5:
  dictstart5[i] = dictstart5.get(i, 0) + 1
dictstart5={k: v for k, v in sorted(dictstart5.items(), key=lambda item: item[1],reverse=True)}
```

```

mid=[x[y] for x in words for y in range(len(x))]
mid3=[x[y:y+3] for x in words3 for y in range(len(x)-2)]
mid4=[x[y:y+4] for x in words4 for y in range(len(x)-3)]
mid5=[x[y:y+5] for x in words5 for y in range(len(x)-4)]

dictmid1 = dict()
for i in mid:
  dictmid1[i] = dictmid1.get(i, 0) + 1
dictmid1={k: v for k, v in sorted(dictmid1.items(), key=lambda item: item[1],reverse=True)}
dictmid3 = dict()
for i in mid3:
  dictmid3[i] = dictmid3.get(i, 0) + 1
dictmid3={k: v for k, v in sorted(dictmid3.items(), key=lambda item: item[1],reverse=True)}
dictmid4 = dict()
for i in mid4:
  dictmid4[i] = dictmid4.get(i, 0) + 1
dictmid4={k: v for k, v in sorted(dictmid4.items(), key=lambda item: item[1],reverse=True)}
dictmid5 = dict()
for i in mid5:
  dictmid5[i] = dictmid5.get(i, 0) + 1
dictmid5={k: v for k, v in sorted(dictmid5.items(), key=lambda item: item[1],reverse=True)}
```

**III. Incidence Matrix**

```
incidence={x:{y:0 for y in alphabet} for x in alphabet}
incidence_probability={x:{y:0 for y in alphabet} for x in alphabet}
letter_occurence={l:len([x for x in words if l in x]) for l in alphabet}
```

```
for i in alphabet:
  incidence[i][i]=len([x for x in words if x.count(i)>1]) #The diagonal counts the number of words with where the letter appears twice
  for j in alphabet:
    if j != i:
      incidence[i][j]=len([x for x in words if ((i in x) and (j in x))])

for i in alphabet:
  for j in alphabet:
    incidence_probability[i][j]=len([x for x in words if ((i in x) and (j in x))])/len([x for x in words if (j in x)])
```

**IV. Palindromes, anagrams, etc**

- Palindromes : words that are the same backwards
- Anagrams : words that share the exact same letters
- Isograms : words that don't contain any letter twice
- Anadromes : words that spell another word backwards

```
palindromes=[x for x in words if x==x[::-1]]
length=max([len(x) for x in palindromes])
longest_palindrome=[x for x in palindromes if len(x)==length]
print('Here is the longest palindrome: '+str(longest_palindrome))
```

```
wordletters=[sorted(list(x)) for x in words]
wordletters_unique=list(sorted(set(map(tuple, wordletters)), reverse=False))
lettersdict={}
for i in range(len(wordletters)):
  lettersdict[tuple(wordletters[i])]=[]
for i in range(len(wordletters)):
  lettersdict[tuple(wordletters[i])].append(words[i])
anagrams={k: v for k, v in sorted(lettersdict.items(), key=lambda item: len(item[1]),reverse=True)}
```

```
longest_anagram=[anagrams[tuple(ii)] for ii in [list(i) for i in list(anagrams.keys()) if len(i) > 21 and len(anagrams[i]) >1]]
print('Here are the longest anagrams: ' + str(longest_anagram))
```

```
#Words with no repeating letter
isograms=[x for x in words if max([list(x).count(i) for i in list(x)])==1]
longest_isogram=[x for x in isograms if len(x)==15]
print('Here is the longest isogram: '+longest_isogram[0]+ '\n And here are some random ones :'+str(isograms[54:68]))
```

```
#Words that spell other words backwards (naturally, it also includes palindromes)
wordsbackwards=[w[::-1] for w in words]
anadromes=[(x,x[::-1]) for x in list(set.intersection(set(words),set(wordsbackwards)))]
print('Here are some examples of anadromes: ' + str(anadromes[:10]))
```
